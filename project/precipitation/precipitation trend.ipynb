```python
import xarray as xr
import pandas as pd

# 读取上传的 NetCDF 文件
file_path = 'rr_ens_mean_0.25deg_reg_2011-2024_v30.0e.nc'  # 文件路径
ds = xr.open_dataset(file_path)

# 查看数据集的结构
print(ds)

# 假设你对 'rr' 变量（降水量）感兴趣，获取这个变量的数据
precipitation = ds['rr']

# 提取 2020-2024 年的数据
precipitation_2020_2024 = precipitation.sel(time=slice('2020-01-01', '2024-12-31'))

# 将数据转换为 Pandas DataFrame
df_2020_2024 = precipitation_2020_2024.to_dataframe().reset_index()

# 查看前几行数据
print(df_2020_2024.head())

# 保存为 CSV 文件
output_csv_path_2020_2024 = '/content/precipitation_2020_2024.csv'
df_2020_2024.to_csv(output_csv_path_2020_2024, index=False)

print(f"CSV file saved for 2020-2024 at: {output_csv_path_2020_2024}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载 CSV 文件
precip_2020_2024 = pd.read_csv('/content/precipitation_2020_2024.csv')

# 查看列名并确认时间列名
print(precip_2020_2024.columns)

# 确保时间列是日期类型（假设时间列名为 'time'）
precip_2020_2024['time'] = pd.to_datetime(precip_2020_2024['time'])

# 提取年份和月份
precip_2020_2024['year'] = precip_2020_2024['time'].dt.year
precip_2020_2024['month'] = precip_2020_2024['time'].dt.month

# 计算每个月的降水总量（按年和月汇总）
monthly_precip_2020_2024 = precip_2020_2024.groupby(['month'])['rr'].sum()

# 输出每个月的总降水量
print("Monthly Total Precipitation (2020-2024):")
print(monthly_precip_2020_2024)

# 绘制2020-2024年每月降水量的时序图
plt.figure(figsize=(10, 6))

# 画出每个月的总降水量
plt.plot(monthly_precip_2020_2024.index, monthly_precip_2020_2024.values, label='2020-2024', color='b', marker='o')

# 设置图表标题和标签
plt.xlabel('Month')
plt.ylabel('Total Precipitation (mm)')
plt.xticks(range(1, 13))  # X轴显示1到12月

# 显示图表
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载2011年和2024年的数据
df_2011 = pd.read_csv('precipitation_2011.csv')
df_2024 = pd.read_csv('precipitation_2024.csv')

# 转换日期列为日期类型
df_2011['time'] = pd.to_datetime(df_2011['time'])
df_2024['time'] = pd.to_datetime(df_2024['time'])

# 设置日期为索引
df_2011.set_index('time', inplace=True)
df_2024.set_index('time', inplace=True)

# 计算前10%的降水量阈值
threshold_2011 = df_2011['rr'].quantile(0.9)
threshold_2024 = df_2024['rr'].quantile(0.9)

print(f"2011 Extreme Precipitation Threshold (90th percentile): {threshold_2011} mm")
print(f"2024 Extreme Precipitation Threshold (90th percentile): {threshold_2024} mm")

# 筛选出降水量超过前10%阈值的事件
extreme_precip_2011 = df_2011[df_2011['rr'] > threshold_2011]
extreme_precip_2024 = df_2024[df_2024['rr'] > threshold_2024]

# 计算每个地点的极端降水频次
extreme_count_2011 = extreme_precip_2011.groupby(['latitude', 'longitude']).size()
extreme_count_2024 = extreme_precip_2024.groupby(['latitude', 'longitude']).size()

# 将频次转换为DataFrame
extreme_count_2011 = extreme_count_2011.reset_index(name='extreme_count_2011')
extreme_count_2024 = extreme_count_2024.reset_index(name='extreme_count_2024')

# 合并2011和2024年的数据
merged_data = pd.merge(extreme_count_2011, extreme_count_2024, on=['latitude', 'longitude'], how='outer').fillna(0)

# 计算每个点的最大频次
merged_data['extreme_count'] = merged_data['extreme_count_2011'] + merged_data['extreme_count_2024']
# 绘制2011年极端降水事件的时空分布图
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['longitude'], merged_data['latitude'], c=merged_data['extreme_count_2011'], cmap='Reds', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Frequency of Extreme Precipitation (2011)')
plt.show()

# 绘制2024年极端降水事件的时空分布图
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['longitude'], merged_data['latitude'], c=merged_data['extreme_count_2024'], cmap='Reds', s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Frequency of Extreme Precipitation (2024)')
plt.show()
```
